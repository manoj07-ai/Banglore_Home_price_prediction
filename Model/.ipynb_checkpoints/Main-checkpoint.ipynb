### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV, RandomizedSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
RANDOM_STATE=42
```

```python
df=pd.read_csv("bengaluru_house_data.csv")
```

### Data Cleaning and Preprocessing

#### Handling Null Values

```python
df.isnull().sum()
```

```python
df.drop(["area_type","availability","society","balcony"],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

#### clean the total_sqft column

```python
df["bhk"]=df["size"].apply(lambda x:int(x.split(" ")[0]))
```

```python
df.drop(["size"],axis=1,inplace=True)
```

```python
df.head()
```

```python
def convert_sqft_to_num(x):
    try:
        x = str(x).lower().strip()
        
        # Case 1: Range (e.g., "2528 - 3188")
        if '-' in x:
            tokens = x.split('-')
            if len(tokens) == 2:
                return (float(tokens[0]) + float(tokens[1])) / 2

        # Case 2: If it contains "sq. meter" or "sqm"
        elif 'meter' in x or 'sq. m' in x:
            value = float(x.split('sq')[0])
            return value * 10.7639  # 1 sq meter = 10.7639 sq ft

        # Case 3: If it contains "sq. yard" or "sq yards"
        elif 'yard' in x or 'sq. yd' in x:
            value = float(x.split('sq')[0])
            return value * 9.0  # 1 sq yard = 9 sq ft

        # Case 4: If it contains "acre"
        elif 'acre' in x:
            value = float(x.split('acre')[0])
            return value * 43560  # 1 acre = 43560 sq ft

        # Case 5: If it contains "perch" 
        elif 'perch' in x:
            value = float(x.split('perch')[0])
            return value * 272.25  # 1 perch = 272.25 sq ft

        # Case 6: Simple number (already in sqft)
        else:
            return float(x)
    
    except:
        return None
```

```python
df['total_sqft']=df["total_sqft"].apply(convert_sqft_to_num)
df=df[df["total_sqft"].notna()]
```

```python
df["price_per_sqft"]=df["price"]*100000/df["total_sqft"]
```

```python
# Group rare locations
location_stats = df['location'].value_counts(ascending=False)
location_less_than_10 = location_stats[location_stats <= 10]
df['location'] = df['location'].apply(lambda x: 'other' if x in location_less_than_10 else x)
```

```python
len(df.location.unique())
```

#### outlier removal

```python
df=df[~((df["total_sqft"]/df["bhk"])<300)]
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df = remove_pps_outliers(df)
 
```

```python
df=remove_pps_outliers(df)
```

```python
df.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2=df[(df["location"]==location) & (df["bhk"]==2)]
    bhk3=df[(df["location"]==location) & (df["bhk"]==3)]
   
    plt.scatter(bhk2.total_sqft,bhk2.price,c="r",label="2bhk")
    plt.scatter(bhk3.total_sqft,bhk3.price,c="b",marker="+",label="2bhk")
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.legend()

plot_scatter_chart(df,"Rajaji Nagar")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])

    for location, location_df in df.groupby("location"):
        bhk_stats = {}

        for bhk, bhk_df in location_df.groupby("bhk"):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df["price_per_sqft"]),
                'std': np.std(bhk_df["price_per_sqft"]),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby("bhk"):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                bad_index = bhk_df[bhk_df["price_per_sqft"] < stats['mean']].index
                exclude_indices = np.append(exclude_indices, bad_index)


    return df.drop(exclude_indices, axis='index')
```

```python
df=remove_bhk_outliers(df)
```

```python
df.shape
```

```python
plot_scatter_chart(df,"Rajaji Nagar")
```

```python
plt.hist(df.price_per_sqft,rwidth=0.8)
plt.xlabel("Price per sqft")
plt.show()
```

```python
df=df[df["bath"]<df["bhk"]+2]
```

```python
df.drop(["price_per_sqft"],axis=1,inplace=True)
```

### OneHotEncoding

```python
df=pd.get_dummies(df, columns=['location'],drop_first=False)
```

```python
df.drop(['location_other'],axis=1,inplace=True)
```

### Split data and train XGBoost

```python
X=df.drop(["price"],axis=1)
```

```python
y=df.price
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=RANDOM_STATE)
```

### Creating Instance of Model

```python
model=XGBRegressor(objective="reg:squarederror",random_state=RANDOM_STATE)
```

```python
param_dist={
    "n_estimators":[64,100,200,300],
    "max_depth":[3,4,5,6,8,10],
    "learning_rate":[0.01,0.5,0.1,0.2],
    "subsample":[0.6,0.7,0.8,1],
    "colsample_bytree":[0.6,0.7,0.8,1.0],
    "gamma":[0,0.1,0.3,0.5]
}
```

### Pre-Training on Random_search_CV

```python
random_search=RandomizedSearchCV(estimator=model,param_distributions=param_dist,n_iter=100,scoring="r2",cv=3,random_state=RANDOM_STATE,verbose=2,n_jobs=-1)
```

```python
random_search.fit(X_train,y_train)
```

```python
print("Best Parameters:",random_search.best_params_)
print("Best Score:",random_search.best_score_)
```

### Finetuning model using GridSearchCV

```python
param_grid = {
    'learning_rate': [0.005, 0.01, 0.02],
    'max_depth': [8, 10, 12],
    'n_estimators': [250, 300, 350],
    'subsample': [0.5, 0.6, 0.7],
    'colsample_bytree': [0.6, 0.7, 0.8],
    'gamma': [0.3, 0.5, 0.7]
}
```

```python
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:",grid_search.best_params_)
print("Best Score:",grid_search.best_score_)
```

### Evaluating Model

```python
best_model=grid_search.best_estimator_
```

```python
y_pred=best_model.predict(X_test)
```

```python
print("Test R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

### Saving the model

```python
import joblib
joblib.dump(best_model,"bangalore_house_price_xgb_model.pkl")
```
